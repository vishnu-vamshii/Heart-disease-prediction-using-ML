### 3.1 K-nearest neighbors

```python
def knn_search(x, D, K, measure):
    if measure == 0:
        dists = np.sqrt(((D - x)**2).sum(axis=1))
    elif measure == 1:
        D_norm = np.array([np.linalg.norm(D[i]) for i in range(len(D))])
        x_norm = np.linalg.norm(x)
        sims = np.dot(D,x)/(D_norm * x_norm)
        dists = 1 - sims
    idx = np.argsort(dists) 
    return idx[:K], dists
```

```python
neigh_idx, distances = knn_search(X_test_norm[0], X_train_norm, 5, 0)
```

```python
print(neigh_idx)
print("\nNearest Neigbors:")
X_train.iloc[neigh_idx]
```

```python
print("Distance of the neighbors to the test instance: ", distances[neigh_idx])
```

```python
#Now that we know the nearest neighbors, we need to find the majority class label among them.
from collections import Counter

print(Counter(y_train[neigh_idx]))
```

```python
def knn_classify(x, D, K, labels, measure):
    from collections import Counter
    neigh_idx, distances = knn_search(x, D, K, measure)
    neigh_labels = labels[neigh_idx]
    count = Counter(neigh_labels)
    predicted_label = count.most_common(1)[0][0]
    return neigh_idx, predicted_label
```

```python
def knn_evaluate(test, test_labs, train, train_labs, K, measure):
    T=0 
    F=0 
    for i in range(len(test)):
        actual=test_labs[i]
        top_K_neighbors, predicted = knn_classify(test[i], train, K, train_labs, measure)
        if actual == predicted:
            T += 1       
        else:
            F += 1     
    accuracy = float(T)/float(T+F) 
    return accuracy
```

```python
import pylab as pl

%matplotlib inline
def plot_coefficients(model, n_features, feature_names, figsize=(20, 50)):
    pl.barh(range(n_features), model.coef_, align='center')
    pl.yticks(np.arange(n_features), feature_names)
    pl.xlabel("Coefficient Value")
    pl.ylabel("Feature")
    pl.ylim(-1, n_features)
    
plot_coefficients(linreg, len(X_train.columns), X_train.columns)
```

```python
def cross_validate(model, X, y, n, verbose=False):
    
    from sklearn.metrics import mean_squared_error
    
    from sklearn.model_selection import KFold
    
    kf = KFold(n_splits=n, random_state=22,shuffle=True) 
    xval_err = 0
    f = 1
    for train,test in kf.split(X):
        model.fit(X[train],y[train]) 
        p_test = model.predict(X[test]) 
        rmse_test = np.sqrt(mean_squared_error(p_test, y[test]))
        if verbose:
            print("Fold %2d RMSE: %.4f" % (f, rmse_test))
        xval_err += rmse_test
        f += 1
    return xval_err/n
```

```python
rmse_10cv = cross_validate(linreg, X_train_norm, y_train, 10, verbose=True)
```

```python
def plot_silhouettes(data, clusters, metric='euclidean'):
    
    from matplotlib import cm
    from sklearn.metrics import silhouette_samples

    cluster_labels = np.unique(clusters)
    n_clusters = cluster_labels.shape[0]
    silhouette_vals = metrics.silhouette_samples(data, clusters, metric='euclidean')
    c_ax_lower, c_ax_upper = 0, 0
    cticks = []
    for i, k in enumerate(cluster_labels):
        c_silhouette_vals = silhouette_vals[clusters == k]
        c_silhouette_vals.sort()
        c_ax_upper += len(c_silhouette_vals)
        color = cm.jet(float(i) / n_clusters)
        pl.barh(range(c_ax_lower, c_ax_upper), c_silhouette_vals, height=1.0, 
                      edgecolor='none', color=color)

        cticks.append((c_ax_lower + c_ax_upper) / 2)
        c_ax_lower += len(c_silhouette_vals)
    
    silhouette_avg = np.mean(silhouette_vals)
    pl.axvline(silhouette_avg, color="red", linestyle="--") 

    pl.yticks(cticks, cluster_labels)
    pl.ylabel('Cluster')
    pl.xlabel('Silhouette coefficient')

    pl.tight_layout()
    pl.show()
    
    return
```

```python
plot_silhouettes(X_train_norm, train_clusters)
```

KMeans clustering is applied to the normalized combined dataset with specified number of clusters i.e., 3. The model is trained to idenEfy clusters in the data and is used to predict the cluster assignments for the instances in the training data and the tesEng data. These predicted cluster assignments were further used in measuring the quality of clustering by compuEng the silhoue,e values for each instance in the data.
